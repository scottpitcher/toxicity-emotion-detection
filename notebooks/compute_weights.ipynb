```python
import pandas as pd
import torch
import ast

# --- Load your processed toxicity CSV ---
df = pd.read_csv("../data/processed/toxicity_raw_processed.csv")

# --- Parse label_encoded from string → list[int] ---
parsed = []
for x in df["label_encoded"]:
    if isinstance(x, str):
        parsed.append(ast.literal_eval(x))
    else:
        parsed.append(x)

# --- Convert list[list[int]] → tensor (N, 6) ---
label_matrix = torch.tensor(parsed, dtype=torch.float32)

print("Label matrix shape:", label_matrix.shape)
print()

# --- Count positives per class ---
pos_counts = label_matrix.sum(dim=0)
print("Positive counts per label:", pos_counts.tolist())
print()

# --- Compute pos_weight = (N - pos) / pos ---
N = label_matrix.shape[0]
pos_weight = (N - pos_counts) / (pos_counts + 1e-6)
print("pos_weight:", pos_weight.tolist())
```

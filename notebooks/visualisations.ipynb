```python
import torch

# Check what's in the checkpoint
checkpoint = torch.load("../models/bert_base_w.pt", map_location="cpu")

print("Keys in checkpoint:")
for key in checkpoint.keys():
    print(f"  {key}")

print("\nKeys in model_state_dict:")
for key in checkpoint['model_state_dict'].keys():
    print(f"  {key}")
    
print(f"\nTotal keys in state_dict: {len(checkpoint['model_state_dict'])}")
```
